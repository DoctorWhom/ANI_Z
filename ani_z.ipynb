```python
%matplotlib inline

import numpy
import math
import matplotlib.pyplot as plot
from mpl_toolkits.mplot3d import Axes3D # used in 3D-plot: projection='3d'

# Hilfsfunktion zur Auswahl der Eingabeverzögerung
def mutual_information(x, x_T):
    x_mat = numpy.concatenate((numpy.matrix(x), numpy.matrix(x_T)), axis=0)    
    
    # Histogramm
    p = numpy.zeros((10, 10))
    N = x_mat.shape[1]
    for i in range(N):
        bin_x0 = min(max(int(x_mat[0, i] * 10), 0), 9) 
        bin_x1 = min(max(int(x_mat[1, i] * 10), 0), 9)
        p[bin_x0, bin_x1] += 1
    p = p / float(N) # normalize (sum probability = 1)
    
    p_i = numpy.sum(p, axis=1)
    p_j = numpy.sum(p, axis=0)
    
    # Mutual Information
    I = 0
    for i in range(10):
        for j in range(10):
            if p[i, j] > 0:
                I += p[i, j] * math.log(p[i, j] / float(p_i[i] * p_j[j]), 2)
    
    return I
    
# Hilfsfunktion zum Zeichnen der Lösung
def plot_2D(x_1, x_2):
    """Visualisierung der Einbettung für 2 Dimensionen"""
    plot.plot(x_1, x_2, 'b.')
    plot.title('2D-Einbettung')
    plot.xlabel('$x$')
    plot.ylabel('$x - T$')
    plot.show()

# Hilfsfunktion zum Zeichnen der Lösung
def plot_3D(x_1, x_2, x_3):
    """Visualisierung der Einbettung für 3 Dimensionen"""
    fig = plot.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x_1, x_2, x_3, 'b.')
    plot.title('3D-Einbettung')
    ax.set_xlabel('$x$')
    ax.set_ylabel('$x - T$')
    ax.set_zlabel('$x - 2 * T$')
    plot.show()
    
# gegeben
t = numpy.arange(0, 20.1, 0.1)
x = 0.4 * numpy.sin(t) + 0.25 * numpy.sin(2 * t + 1.5) + 0.65

# Visualisierung
plot.plot(t, x, 'b.')
plot.title('zeitliches Signal')
plot.xlabel('$t$')
plot.ylabel('$x$')
plot.show()

# Zahlen ersetzen durch Ihre berechneten Werte
T = 0.0
d = 2

# Lösung berechnen und Visualisierung
xs = []
for i in range(d):
    xi = 0.4 * numpy.sin(t - i * T) + 0.25 * numpy.sin(2 * (t - i * T) + 1.5) + 0.65
    xs.append(xi)
if d == 2:
    plot_2D(xs[0], xs[1])
else:
    plot_3D(xs[0], xs[1], xs[2])

result_x = []
for i in range(d):
    result_x.append(numpy.matrix(xs[i]))
result = numpy.concatenate(result_x, axis=0)

# Ergebnisse überprüfen
import test_z_ip
result_x = []
for i in range(d):
    result_x.append(numpy.matrix(xs[i]))
result = numpy.concatenate(result_x, axis=0)
test_z_ip.loesung_z9(result)
```
